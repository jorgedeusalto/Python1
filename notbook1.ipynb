```python
from PyPDF2 import PdfFileMerger, PdfFileReader
import glob, os
path = "C:\\Users\\jorge\\Desktop\\Pdf_temp\\DWG2PDF\\"
os.system('cls' if os.name == 'nt' else 'clear')
os.chdir(path)
```

```python
list1 = []#list all file on the directory
for file in glob.glob("*.pdf"):
    list1.append(file)
    #print(file)
```

```python
list1.sort(key=lambda x: x.split("-", 1)[-1])
print("LIST1 is  >>>>  ")
print(list1)
print('>>>>  end list <<<<<<')
print('There are ' + str(len(list1)) + ' itens in LIST1')
```

```python
list2=[]
for x1 in list1:
    x2 = x1.find('-')# get first position of '-'
    x3 = (x1[x2+1:])# store the rest of the string from position stored in x2
    x4 = x3.find('-')# get second position for '-'
    x5 = x2+x4+1#get position of second '-' in total string
    x6 = (x1[:x5])#storage string part up to second '-'
    list2.append(x6)# append item to list2
list2.sort(key=lambda x: x.split("-", 2)[-1])
print("LIST2 is  >>>>  ")
print(list2)
print('>>>>  end list <<<<<<')
print('There are ' + str(len(list2)) + ' itens in LIST2')
```

```python
list3 = set()# set just storage unique values
for x7 in list2:
    list3.add(x7)
list3 = list(list3)#convert set to list
list3.sort(key=lambda x: x.split("-", 2)[-1])
print("LIST3 is  >>>>  ")
print(list3)
print('>>>>  end list <<<<<<')
print('There are ' + str(len(list3)) + ' itens in LIST3')
```
